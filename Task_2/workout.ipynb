```python
#option with resizing
import cv2
import imutils
import numpy as np

# # Read the input images
image1 = cv2.imread(r"D:\desktop\SANDEEP SIR\sample_project_for_banao_ML\Task_2\input1.png")
image2 = cv2.imread(r"D:\desktop\SANDEEP SIR\sample_project_for_banao_ML\Task_2\input2.png")

def resize_images(image1, image2):
    # Get the dimensions of both images
    h1, w1 = image1.shape[:2]
    h2, w2 = image2.shape[:2]

    # Resize the larger image to match the dimensions of the smaller image
    if h1 * w1 > h2 * w2:
        image1_resized = cv2.resize(image1, (w2, h2))
        image2_resized = image2
    else:
        image1_resized = image1
        image2_resized = cv2.resize(image2, (w1, h1))

    return image1_resized, image2_resized

# # Read the input images
# image1 = cv2.imread(r"D:\desktop\Spot-6-Differences-part 1.jpg")
# image2 = cv2.imread(r"D:\desktop\Spot-6-Differences-part 2.jpg")
cv2.imshow('INPUT Image 1', image1)
cv2.imshow('INPUT Image 2', image2)
cv2.waitKey(0)
cv2.destroyAllWindows()


image1resize,image2resize = resize_images(image1, image2)




# Convert images to grayscale
gray1 = cv2.cvtColor(image1resize, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2resize, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blurring to reduce noise
gray1 = cv2.GaussianBlur(gray1, (5, 5), 0)
gray2 = cv2.GaussianBlur(gray2, (5, 5), 0)

# Calculate the absolute difference between the grayscale images
diff = cv2.absdiff(gray1, gray2)

# Apply thresholding to highlight the differences
thresh = cv2.threshold(diff, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# Dilation to enhance the differences
kernel = np.ones((5, 5), np.uint8)
dilate = cv2.dilate(thresh, kernel, iterations=2)

# Finding contours
contours = cv2.findContours(dilate.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)

# # Display the difference image
# cv2.imshow('Differences', dilate)
# cv2.waitKey(0)
# cv2.destroyAllWindows()


# Draw bounding boxes on the original images
for contour in contours:
    if cv2.contourArea(contour) > 100:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image1, (x, y), (x + w, y + h), (0, 0, 255), 2)
        cv2.rectangle(image2, (x, y), (x + w, y + h), (0, 0, 255), 2)

        
        
        
def calculate_similarity_score(image1, image2):
    

    # Calculate the number of non-zero pixels (differences) in the difference image
    diff_pixels = np.count_nonzero(diff)

    # Calculate the total number of pixels in the images
    total_pixels = image1.shape[0] * image1.shape[1]

    # Calculate the similarity score as the percentage of similar pixels
    similarity_score = ((total_pixels - diff_pixels) / total_pixels) * 100

    return similarity_score





# Display the original images with bounding boxes
cv2.imshow('Image 2 with Bounding Boxes', image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite("D:\desktop\SANDEEP SIR\sample_project_for_banao_ML\Task_2\output.png",image2)
print('similarity score is',calculate_similarity_score(image1, image2))
```
